## ë°ì´í„°ì˜ ê°œìˆ˜

````python
x = np.array([
    18, 5, 10, 23, 19, -8, 10, 0, 0, 5, 2, 15, 8, 
    2, 5, 4, 15, -1, 4, -7, -24, 7, 9, -6, 23, -13
])
x
````

## í‘œë³¸ í‰ê· 
* ìš°ë¦¬ê°€ ì¼ë°˜ì ìœ¼ë¡œ ì•„ëŠ” í‰ê· 
* í†µê³„ìš©ì–´ë¡œëŠ” í‘œë³¸ í‰ê· (sample average, sample mean)
<br>
$
\bar{x} = \frac{1}{N}\displaystyle\sum_{i=1}^{N}{x_i}
$
(ğ‘ì€ ë°ì´í„°ì˜ ê°œìˆ˜)

````python
np.mean(x), x.mean()
````

## í‘œë³¸ ë¶„ì‚°
* í‘œë³¸ ë¶„ì‚°(sample variance) : ë°ì´í„°ì™€ í‘œë³¸ í‰ê· ê°„ì˜ ê±°ë¦¬ì˜ ì œê³±ì˜ í‰ê· 
* í‘œë³¸ ë¶„ì‚°ì´ ì‘ìœ¼ë©´ ë°ì´í„°ê°€ ëª¨ì—¬ìˆëŠ” ê²ƒì´ê³  í¬ë©´ í©ì–´ì ¸ ìˆëŠ” ê²ƒ
<br>
$
s^2 = \frac{1}{N}\displaystyle\sum_{i=1}^{N}{(x_i-\bar{x})^2}
$

````python
np.var(x)
````

## í‘œë³¸ í‘œì¤€í¸ì°¨
* í‘œë³¸ í‘œì¤€í¸ì°¨(sample standard variance) : í‘œë³¸ ë¶„ì‚°ì˜ ì–‘ì˜ ì œê³±ê·¼ ê°’ 
<br>
$
s = \sqrt{s^2}
$

````python
np.std(x)
````
