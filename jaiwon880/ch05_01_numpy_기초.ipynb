* ë°°ì—´ ê°ì²´ íƒ€ì…(ìë£Œí˜•)
* ë¦¬ìŠ¤íŠ¸ì™€ ë™ì¼í•´ ë³´ì´ì§€ë§Œ ì°¨ì´ê°€ ë§ìŒ

|ìë£Œí˜•|íŠ¹ì§•|
|:-|:-|
|ë¦¬ìŠ¤íŠ¸(list)|ê°ê°ì˜ ì›ì†Œê°€ ë‹¤ë¥¸ ìë£Œí˜•ì´ ë  ìˆ˜ ìˆìŒ|
|ë°°ì—´(array)|ì—°ì†ì ì¸ ë©”ëª¨ë¦¬ ë°°ì¹˜ë¥¼ ê°€ì§€ê¸° ë•Œë¬¸ì— ëª¨ë“  ì›ì†Œê°€ ê°™ì€ ìë£Œí˜•ì´ì–´ì•¼ í•¨<br>ì›ì†Œì— ëŒ€í•œ ì ‘ê·¼ê³¼ ë°˜ë³µë¬¸ ì‹¤í–‰ì´ ë¹¨ë¼ì§|

````python
import random
def average_python(n):

  '''
  nê°œì˜ ëœë¤í•œ 0~1ê°œì˜ ì‹¤ìˆ˜ë¥¼ í‰ê· í•´ì£¼ëŠ” í•¨ìˆ˜
  '''
  
  s = 0
  for i in range(n):
    s += random.random()
    return s/n
````

````python
%time average_python(100_000)
````

````python
%timeit average_python(100_000) # ì—¬ëŸ¬ë²ˆ ë°˜ë³µ(loop) í•˜ê¸°
````

````python
def average_python(n):
  'nê°œì˜ ëœë¤í•œ 0~1ê°œì˜ ì‹¤ìˆ˜ë¥¼ í‰ê· í•´ì£¼ëŠ” í•¨ìˆ˜'
  s = np.random.random(n)
  return s.mean()
````

````python
%time average_python(100_000)
````

````python
%timeit average_python(100_000) # ì—¬ëŸ¬ë²ˆ ë°˜ë³µ(loop) í•˜ê¸°
````

````python
import numpy as np

arr = np.array(range(1,6))
x = arr.copy()
arr[0] = 42
print(f"arr: {arr}")
print(f"x: {x}")
````

````python
arr = np.array(range(1,6))
x = arr.view()
arr[0] = 40
print(f"arr: {arr}")
print(f"x: {x}")
````

````python
x = arr.copy()
y = arr.view()

# ë°°ì—´ ë°ì´í„° ì†Œìœ  ì—¬ë¶€ ë°˜í™˜
print(x.base)
print(y.base)
````
